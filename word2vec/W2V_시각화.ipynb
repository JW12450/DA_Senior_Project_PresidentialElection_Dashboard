# 대안 - 후보별 기사 구분하여 학습


```python
news_lee = news[news['후보명'] =='이재명']
news_lee.reset_index(drop=True, inplace = True)
news_yoon = news[news['후보명'] =='윤석열']
news_yoon.reset_index(drop=True, inplace = True)
news_ahn = news[news['후보명'] =='안철수']
news_ahn.reset_index(drop=True, inplace = True)
print(f'이재명 : {len(news_lee)}개, 윤석열 : {len(news_yoon)}개, 안철수 : {len(news_ahn)}개')
```

```python
news_lee = news_lee['전처리최종']
news_yoon = news_yoon['전처리최종']
news_ahn = news_ahn['전처리최종']
```

```python
lst1 = []
lst2 = []
lst3 = []

for i in range(len(news_lee)-1):
    lst1.append(news_lee[i].split(" "))
for j in range(len(news_yoon)-1):
    lst2.append(news_yoon[j].split(" "))
for k in range(len(news_ahn)-1):
    lst3.append(news_ahn[k].split(" "))
```

```python
model_lee = Word2Vec(sentences=lst1, size=100, window=5, min_count=5, workers=4, sg=1)
model_yoon = Word2Vec(sentences=lst2, size=100, window=5, min_count=5, workers=4, sg=1)
model_ahn = Word2Vec(sentences=lst3, size=100, window=5, min_count=5, workers=4, sg=1)
```

```python
print(model_lee.wv.most_similar("이재명", topn= 100))
print(model_yoon.wv.most_similar("윤석열", topn= 100))
print(model_ahn.wv.most_similar("안철수", topn= 100))
```

```python
model_lee.wv.save_word2vec_format('model_lee')
model_yoon.wv.save_word2vec_format('model_yoon')
model_ahn.wv.save_word2vec_format('model_ahn')
```
